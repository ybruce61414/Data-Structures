```python
#binary_tree
#in this case, we use one class binarytree() to build binary tree
'''in computer science, a binary tree is a data structure in 
which each node has at the most two children, which are referred
to as the left and the right child.'''

class binarytree():
    def __init__(self,value):
        self.value = value
        self.Lchild = None
        self.Rchild = None
        self.parent = None
                    
    def insertleft(self,value):
        if self.Lchild == None:
            self.Lchild = binarytree(value) 
            #Lchild's parent: point to itself
            self.Lchild.parent = self      
        else:
            newnode = binarytree(value)
            newnode.Lchild = self.Lchild
            self.Lchild = newnode
            #set the parent
            newnode.parent = self
            self.Lchild.parent = newnode
            
    def insertright(self,value):
        if self.Rchild == None:
            self.Rchild = binarytree(value)
            #Rchild's parent: point to itself
            self.Rchild.parent = self  
        else:
            newnode = binarytree(value)
            newnode.Rchild = self.Rchild
            self.Rchild = newnode
            #set the parent
            newnode.parent = self
            self.Rchild.parent = newnode

    #Depth First Traversals
    def preorder(self):
        #root-left-right
        print(self.value + ' ',end='')
        if self.Lchild:
            self.Lchild.preorder()
        if self.Rchild:
            self.Rchild.preorder()     
    
    def inorder(self):
        #left-root-right
        if self.Lchild:
            self.Lchild.inorder()
        print(self.value + ' ',end='') 
        if self.Rchild:
            self.Rchild.inorder()
            
    def postorder(self):
        #left-right-root
        if self.Lchild:
            self.Lchild.postorder()
        if self.Rchild:
            self.Rchild.postorder()        
        print(self.value + ' ',end='')  
    
    #----- using level order traversal ------
def plot(rootnode):
    import pydot 
    graph = pydot.Dot(graph_type='graph',label='Binary Search Tree')
    graph.set_node_defaults(shape='ellipse',fontsize=15,
                            height=.45, width=.85,
                            style='filled', fillcolor='gray87')    
    #empty quene for level order traversal
    quene = []
    quene.append(rootnode)
    
    while len(quene) > 0:
        #add new node
        new_node = pydot.Node(name= str(quene[0].value))
        graph.add_node(new_node) 
        #set parent node (if the new node not the root)
        if quene[0].parent:
            new_node_par = pydot.Node(name= str(quene[0].parent.value))
            graph.add_node(new_node_par) 
            graph.add_edge(pydot.Edge(new_node_par,new_node,
                                      dir ="forward",arrowhead='inv',penwidth=1))     
        node = quene.pop(0)
        #enquene left child
        if node.Lchild:
            quene.append(node.Lchild)
        #enquene right child
        if node.Rchild:
            quene.append(node.Rchild)    
            
    #setting graph            
    graph.write_png('binarytreetest.png')
    import IPython.display as dis
    image = dis.Image('binarytreetest.png')
    dis.display(image)        
#---------------------------------------- 

anode = binarytree('A')
anode.insertleft('B')
anode.insertright('C')

bnode = anode.Lchild
bnode.insertleft('D')
bnode.insertright('E')

cnode = anode.Rchild
cnode.insertright('F')

anode.preorder()
print('-- Preorder')
anode.inorder()
print('-- Inorder')
anode.postorder()
print('-- Postorder')

plot(anode)
```
