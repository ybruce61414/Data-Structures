```python
#binary_search_tree

class treenode():
    def __init__(self,value):
        self.value = value
        self.Lchild = None
        self.Rchild = None        

class binary_search_tree():
    def __init__(self):
        self.root = None
        
    def insert(self,value):
        if self.root == None:
            self.root = treenode(value)
        else:
            self.insert_rec(self.root,value)
    
    def insert_rec(self,cur_node,value):
        #for calling recursive function
        if value < cur_node.value:
            if cur_node.Lchild:
                self.insert_rec(cur_node.Lchild,value)  
            else:
                cur_node.Lchild = treenode(value)
        elif value > cur_node.value: 
            if cur_node.Rchild:
                self.insert_rec(cur_node.Rchild,value)    
            else:
                cur_node.Rchild = treenode(value)
        elif value == cur_node.value:
            print('value: '+str(value)+' has already existed in the tree.')
            
    def preorder(self):
        #root-left-right
        if self.root != None:
            self.preorder_rec(self.root)
   
    def preorder_rec(self,cur_node):
        #for calling recursive function
        print(str(cur_node.value) + ' ',end='')
        if cur_node.Lchild:
            self.preorder_rec(cur_node.Lchild)
        if cur_node.Rchild:
            self.preorder_rec(cur_node.Rchild)  
        
    #----- tree visualization ------
    #----- only for this case ------
    def plot(self):
        import pydot 
        graph = pydot.Dot(graph_type='graph',label='Binary Search Tree')
        L2_list = [4,16]
        LL3_list, LR3_list = [1,5], [8,'None ']
        LL4_list, LR4_list = [' None',2], [' None ',11]
        LR5_list = [9,12]
        LR6_list = ['None  ',15] 
        #level 1,2 
        for i in L2_list:
            edge = pydot.Edge(str(7), "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge)
        
        #level 3 left subtree  
        for i in LL3_list:
            edge = pydot.Edge(str(4), "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge)      
        #level 3 right subtree  
        for i in LR3_list:
            edge = pydot.Edge(str(16), "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge) 
        
        #level 4 left subtree  
        for i in LL4_list:
            edge = pydot.Edge(str(1), "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge)  
        #level 4 right subtree  
        for i in LR4_list:
            edge = pydot.Edge(str(8), "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge)
        
        #level 5 right subtree  
        for i in LR5_list:
            edge = pydot.Edge(str(11), "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge)
        #level 6 right subtree  
        for i in LR6_list:
            edge = pydot.Edge(str(12), "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge)    
        
        #setting graph            
        graph.write_png('binarytree.png')
        import IPython.display as dis
        image = dis.Image('binarytree.png')
        dis.display(image)        
#----------------------------------------         
        
a =  binary_search_tree()       
list = [7,4,1,5,16,8,11,12,15,9,2,2]
for i in list:
    a.insert(i)
#use preorder to check if the BST is correct
a.preorder()
print('--- Preorder')
a.plot()
```
