```python
#binary_tree
#in this case, we use one class binarytree() to build binary tree
'''in computer science, a binary tree is a data structure in 
which each node has at the most two children, which are referred
to as the left and the right child.'''


class binarytree():
    def __init__(self,value):
        self.value = value
        self.Lchild = None
        self.Rchild = None
            
    def insertleft(self,value):
        if self.Lchild == None:
            self.Lchild = binarytree(value)
        else:
            newnode = binarytree(value)
            newnode.Lchild = self.Lchild
            self.Lchild = newnode
    
    def insertright(self,value):
        if self.Rchild == None:
            self.Rchild = binarytree(value)
        else:
            newnode = binarytree(value)
            newnode.Rchild = self.Rchild
            self.Rchild = newnode

    #tree traversals
    def preorder(self):
        #root-left-right
        print(self.value + ' ',end='')
        if self.Lchild:
            self.Lchild.preorder()
        if self.Rchild:
            self.Rchild.preorder()     
    
    def inorder(self):
        #left-root-right
        if self.Lchild:
            self.Lchild.inorder()
        print(self.value + ' ',end='') 
        if self.Rchild:
            self.Rchild.inorder()
            
    def postorder(self):
        #left-right-root
        if self.Lchild:
            self.Lchild.postorder()
        if self.Rchild:
            self.Rchild.postorder()        
        print(self.value + ' ',end='')  
        
    #---------------------------------------- 
    def plot(self):
        #tree visualization
        #only for this case 
        import pydot 
        graph = pydot.Dot(graph_type='graph',label='Binary Tree')
        L1_list = [self.Lchild.value,self.Rchild.value]
        L2_list = [self.Lchild.Lchild.value,self.Lchild.Rchild.value]
        
        #level 1,2 
        for i in L1_list:
            edge = pydot.Edge(self.value, "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge)
        
        #level 3 left subtree  
        for i in L2_list:
            edge = pydot.Edge(self.Lchild.value, "%s" % i, dir ="forward",arrowhead='inv')
            graph.add_edge(edge)  
        
        #level 3 right subtree 
        node_em = pydot.Node("None",style="filled", fillcolor="gray87")
        node_f = pydot.Node(self.Rchild.Rchild.value)
        graph.add_node(node_em)
        graph.add_node(node_f)
        graph.add_edge(pydot.Edge(self.Rchild.value, node_em,dir ="forward",arrowhead='inv'))
        graph.add_edge(pydot.Edge(self.Rchild.value, node_f,dir ="forward",arrowhead='inv'))
        
        #setting graph            
        graph.write_png('binarytree.png')
        import IPython.display as dis
        image = dis.Image('binarytree.png')
        dis.display(image)        
            
#---------------------------------------- 

anode = binarytree('A')
anode.insertleft('B')
anode.insertright('C')

bnode = anode.Lchild
bnode.insertleft('D')
bnode.insertright('E')

cnode = anode.Rchild
cnode.insertright('F')

anode.preorder()
print('-- Preorder')
anode.inorder()
print('-- Inorder')
anode.postorder()
print('-- Postorder')
anode.plot()
```
