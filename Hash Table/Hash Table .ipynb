```python
#map abstract data type

class hash_table():
    def __init__(self):
        self.capacity = 11
        self.slots = [{} for i in range(self.capacity)]
        
    def showall(self):
        for i in range(self.capacity):
            print('hash: '+ str(i), str(self.slots[i]))
        return 
    #-------------------------------------
    
    def hashfunction(self,key,capacity):
        return key % capacity
    
    def rehash(self,pre_hash,capacity):
        return (pre_hash + 1) % capacity
    
    #-------------------------------------
    def add(self,key,data):
        hashvalue = self.hashfunction(key,self.capacity)
        
        if not self.slots[hashvalue]:
            self.slots[hashvalue] = {key:data}
        
        elif self.slots[hashvalue].keys() == {key}:
            #same key,repalce
            self.slots[hashvalue] = {key:data}
        
        else:
            #open addressing (linear probing)
            nexthash = self.rehash(hashvalue,self.capacity)
            
            while self.slots[nexthash] and \
            self.slots[nexthash].keys() != {key}:
                nexthash = self.rehash(nexthash,self.capacity)
            
            if not self.slots[nexthash]:
                self.slots[nexthash] = {key:data}
            
            else:
                #same key,replace
                self.slots[hashvalue] = {key:data}
                
    #------------------------------------- 
    def search(self,key):
        hashvalue = self.hashfunction(key,self.capacity)
        
        if self.slots[hashvalue].keys() == {key}:
            return self.slots[hashvalue][key]
        
        else:
            #sequential search
            nexthash = self.rehash(hashvalue,self.capacity)
            
            while not self.slots[nexthash] or \
            self.slots[nexthash].keys() != {key}:
                
                if nexthash == hashvalue:
                    #encountering starting point,stop
                    return print('not found')
                else:
                    nexthash = self.rehash(nexthash,self.capacity)      
            
            return self.slots[nexthash][key]    

#------------------testing------------------- 
a = hash_table()
a.add(100,'bruce')
a.add(98,'irene')
a.add(95,'rechard')
a.add(12,'m')      
a.add(23,'coco')
a.add(106,'alex')
a.showall()   
a.search(106)
a.search(3)
```
