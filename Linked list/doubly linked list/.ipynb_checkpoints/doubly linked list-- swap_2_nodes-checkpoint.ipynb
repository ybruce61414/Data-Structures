```python
#singly doubly linked lists 
import sys as sys    

class student():
    def __init__(self,name,number,math,eng):
        self.name = name
        self.math = math
        self.eng = eng
        self.no = number
        self.next = None
        self.prev = None
        
    def __str__(self):
        return "NO.%s: %s\tMath: %d\tEnglish: %d" % (self.no,self.name,self.math,self.eng) 
    
class DoublyLinkedList():
    def __init__(self):
        self.head = None
        
    def append(self,name,number,math,eng):
        new_one = student(name,number,math,eng)
        current = self.head
        
        if not self.head:                               # no element in the list
            self.head = new_one   #its pointer next and previous are already point to None
        else:
            while current.next:
                current = current.next
            current.next = new_one  
            new_one.next = None
            new_one.prev = current
            
    def prepend(self,name,number,math,eng):
        new_one = student(name,number,math,eng)
        current = self.head
        
        if not self.head:
            self.head = new_one
        else:
            current.prev = new_one    
            new_one.next = current
            self.head = new_one
            
    def swap_2nodes(self,name1,name2):
        current1,current2 = self.head,self.head
        prev1,prev2 = None,None
        
        if name1 == name2:
            print('same person')
            return
        while current1 and current1.name != name1:
            prev1 = current1
            current1 = current1.next
        while current2 and current2.name != name2:
            prev2 = current2
            current2 = current2.next    
        
        if not current1 or not current2:
            print('Either %s or %s is not in the list.' % (name1,name2))
            sys.exit()
            
        if prev1 == None:
            self.head = current2
            current2.prev = None
        else:
            prev1.next = current2
            current2.prev = None
        
        if prev2 == None:
            self.head = current1
            current1.prev = None
        else:
            prev2.next = current1
            current1.prev = None
            
        temp = current1.next
        current1.next = current2.next
        current2.next = temp                       
     
    def addafter_node(self,afternode,name,number,math,eng):
        pass
    def addbefore_node(self,beforenode,name,number,math,eng):
        pass
    def remove(self,name):       
        pass
    def reverse(self):
        pass
    def printlist(self):
        current = self.head
        while current:
            print(current)
            current = current.next
            
#-----------------------------------------------        
dlist = DoublyLinkedList()
dlist.append('viola',1,88,100)
dlist.append('irene',2,67,89)
dlist.append('andy',3,45,67)
dlist.swap_2nodes('irene','andy')

dlist.printlist()
```
