```python
#circular linked list 
import sys as sys

class student():
    def __init__(self,name,number,math,eng):
        self.name = name
        self.math = math
        self.eng = eng
        self.no = number
        self.next = None

    def __str__(self):
        return "NO.%d: %s\tMath: %d\tEnglish: %d" % (self.no,self.name,self.math,self.eng)    

class CircularList():
    def __init__(self):
        self.head = None
        
    def append(self,name,number,math,eng):            
        #add item to the last
        if not self.head:
            self.head = student(name,number,math,eng)
            self.head.next = self.head
        else:
            new_one = student(name,number,math,eng)
            current = self.head
            
            while current.next != self.head:
                current = current.next
            current.next = new_one
            new_one.next = self.head
            
    def prepend(self,name,number,math,eng):          
        #add item to the front
        new_one = student(name,number,math,eng)
        new_one.next = self.head
        current = self.head
        
        if not self.head:                            
            #when self.head is None, no element in the list
            new_one.next = new_one
        else:
            while current.next != self.head:
                current = current.next
            current.next = new_one
        self.head = new_one    
    
    def swap_2nodes(self,name1,name2):            
        current1,current2 = self.head,self.head
        prev1,prev2 = None,None
        
        if name1 == name2:
            print('same person')
            return
        while current1 and current1.name != name1: 
            prev1 = current1
            current1 = current1.next
            if current1 == self.head:
                current1 = None
                break
        while current2 and current2.name != name2:
            prev2 = current2
            current2 = current2.next
            if current2 == self.head:
                current2 = None
                break
        
        if not current1 or not current2:
            print('Either %s or %s is not in the list.' % (name1,name2))
            sys.exit()
                               
        if current1 == self.head:                  
            #find prev node
            temp = self.head
            while temp and temp.next != self.head:
                temp = temp.next
            prev1 = temp    
            self.head = current2
        
        elif current2 == self.head:
            temp = self.head
            while temp and temp.next != self.head:
                temp = temp.next
            prev2 = temp    
            self.head = current1
        
        prev1.next, prev2.next = current2, current1    
            
        temp = current1.next
        current1.next = current2.next
        current2.next = temp           
    
    def remove(self,name):
        pass
    def size(self):
        pass
    def printlist(self):
        current = self.head
        while current:
            print(current)
            current = current.next
            if current == self.head:
                break
                
                
list = CircularList()     
list.append('viola',1,99,100)
list.append('irine',2,67,89)
list.append('andy',3,56,78)
list.swap_2nodes('viola','andy')

list.printlist()
```
