```python
#circular linked list 

class student():
    def __init__(self,name,number,math,eng):
        self.name = name
        self.math = math
        self.eng = eng
        self.no = number
        self.next = None

    def __str__(self):
        return "NO.%d: %s\tMath: %d\tEnglish: %d" % (self.no,self.name,self.math,self.eng)    

class CircularList():
    def __init__(self):
        self.head = None
        
    def append(self,name,number,math,eng):            
        #add item to the rear
        if not self.head:
            self.head = student(name,number,math,eng)
            self.head.next = self.head
        else:
            new_one = student(name,number,math,eng)
            current = self.head
            
            while current.next != self.head:
                current = current.next
            current.next = new_one
            new_one.next = self.head
            
    def prepend(self,name,number,math,eng):          
        #add item to the front
        new_one = student(name,number,math,eng)
        new_one.next = self.head
        current = self.head
        
        if not self.head:                            
            #when self.head is None, no element in the list
            new_one.next = new_one
        else:
            while current.next != self.head:
                current = current.next
            current.next = new_one
        self.head = new_one    
        
    
    def addafter_node(self,afternode,name,number,math,eng):
        current = self.head
        while current:
            if current.next == self.head and current.name == afternode:          
                #add after the last one
                self.append(name,number,math,eng)
                return
            elif current.next == self.head and current.name != afternode:        
                #name is not in the list 
                print('\nThe person after whom you wanna add isn\'t in the list\n')
                return 
            elif current.name == afternode:
                new_one = student(name,number,math,eng) 
                new_one.next = current.next
                current.next = new_one
                return
            current = current.next      
            
    def addbefore_node(self,beforenode,name,number,math,eng):
        current = self.head 
        prev = student(name,number,math,eng) 
        while current and prev.next != self.head:                          
            #find the prev before head
            prev = current
            current = current.next
            
        while current:
            if prev.next == self.head and current.name == beforenode:
                self.prepend(name,number,math,eng)
                return
            elif current.next == self.head and current.name != beforenode: 
                #name is not in the list
                print('\nThe person before whom you wanna add isn\'t in the list\n')
                return 
            elif current.name == beforenode:
                new_one = student(name,number,math,eng) 
                new_one.next = current
                prev.next = new_one
                return
            prev = current
            current = current.next   
    
            
    def swap_2nodes(self,name1,name2):            
        pass
    def remove(self,name):
        pass
    def size(self):
        pass
    
    def printlist(self):
        current = self.head
        while current:
            print(current)
            current = current.next
            if current == self.head:
                break
                
                
list = CircularList()
list.append('viola',1,99,100)
list.append('irine',2,67,89)
list.append('andy',3,56,78)
list.addafter_node('irine','ikea',4,66,77)
list.addbefore_node('irine','bruce',5,78,85)

list.printlist()
```
